# Stem Plots

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":") # linefmt=":"

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="r+") # markerfmt="r+"

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro") # markerfmt="ro"

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=1) # bottom=2

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=9) # bottom=2

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0) # bottom=0

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g") # basefmt="g"

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="Python") # label="Python" and legend()

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="Python", use_line_collection=True) # use_line_collection=True

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="Python", use_line_collection=False) # use_line_collection=False

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="Python", use_line_collection=False,
        orientation="horizontal") #  orientation="horizontal"

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=":", markerfmt="ro", bottom=0, basefmt="g", label="Python",
        ) #  orientation="horizontal"

plt.legend()
plt.show()
```
