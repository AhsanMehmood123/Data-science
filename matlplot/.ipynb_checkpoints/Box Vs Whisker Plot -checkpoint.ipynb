# Box Vs Whisker Plot 

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,150]

plt.boxplot(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True) # notch = True

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,vert=True) # vert=True

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,vert=False) # vert=False

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.8) # widths=0.8

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"]) # labels=["Python"]

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"], patch_artist=True) # patch_artist=True
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"],showmeans=True) # showmeans=True

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],showmeans=True) # 
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],showmeans=True, whis=2) # whis=2
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g") # sym = "g"
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g+") # sym = "g+"
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"}) # boxprops={"color":"g"}
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"}) # capprops={"color": "r"
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"},
           whiskerprops={"color": "y"}) # whiskerprops={"color": "y"}
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"},
           whiskerprops={"color": "y"},flierprops={"markeredgecolor":"b"}) # whiskerprops={"color": "y"}
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 = [10,20,40,50,30,60,90]

plt.boxplot(x,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"},
           whiskerprops={"color": "y"}) # whiskerprops={"color": "y"}

plt.boxplot(x1,labels=["Python"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"},
           whiskerprops={"color": "y"}) # whiskerprops={"color": "y"}

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 = [10,20,40,50,30,60,90]

y = [x,x1]

plt.boxplot(y,labels=["Python","c"],showmeans=True,sym = "g",boxprops={"color":"g"},capprops={"color": "r"},
           whiskerprops={"color": "y"}) #


plt.show()
```
