# Axis

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)

plt.yticks(y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x,y)

plt.xticks(x, labels=["python", "C", "C++", "Java"])

plt.yticks(y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x,y)

#plt.xticks(x, labels=["python", "C", "C++", "Java"])

#plt.yticks(y)

plt.xlim(0,10)

plt.show()
```
