# Scatter plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
c = ["r", "y", "g", "b", "r", "g", "r" ]
plt.scatter(x,y,color = c)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
c = ["r", "y", "g", "b", "r", "g", "r" ]
size = [10,2,4,5,2,10,7]
plt.scatter(x,y,color = c, s = size)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
c = ["r", "y", "g", "b", "r", "g", "r" ]

plt.scatter(x,y,color = c, alpha =0.5)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
c = ["r", "y", "g", "b", "r", "g", "r" ]
size = [100,200,400,50,200,100,70]
plt.scatter(x,y,color = c,s=size, marker = "*", edgecolors="g")
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
c = ["r", "y", "g", "b", "r", "g", "r" ]
size = [100,200,400,50,200,100,70]
plt.scatter(x,y,color = c,s=size, marker = "*", edgecolors="g", linewidths=1)
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
color = [10,49,30,29,56,20,30]
size = [100,200,400,50,200,100,700]
plt.scatter(x,y, c = color, s=size, cmap='viridis')

plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
color = [10,49,30,29,56,20,30]
size = [100,200,400,50,200,100,700]
plt.scatter(x,y, c = color, s=size, cmap='viridis')
plt.colorbar()
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
color = [10,49,30,29,56,20,30]
size = [100,200,400,50,200,100,700]
plt.scatter(x,y, c = color, s=size, cmap='rainbow', alpha=0.5)
t = plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
z = [3,2,4,5,1,6,2]
color = [10,49,30,29,56,20,30]
size = [100,200,400,50,200,100,700]
plt.scatter(x,y, c = color, s=size, cmap='rainbow', alpha=0.5)
plt.scatter(x,z, c = color, s=size, alpha=0.5)
t = plt.colorbar()
t.set_label("Color Bar")
plt.title("Scatter plot", fontsize=15)
plt.xlabel("Day")
plt.ylabel("number")
plt.show()
```
