# Area Vs Stack Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]

plt.stackplot(x,area)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [2,3,2,4,2]

plt.stackplot(x,area,area1,area2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [2,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area,area1,area2,labels=l, colors=["r","g","b"])

plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [2,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area,area1,area2,labels=l, colors=["r","g","b"],baseline="zero") # baseline="zero"

plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [2,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area,area1,area2,labels=l, colors=["r","g","b"],baseline="sym") # baseline="sym"

plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
area = [2,3,2,5,4]
area1 = [2,3,4,5,6]
area2 = [2,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area,area1,area2,labels=l, colors=["r","g","b"],baseline="wiggle") # baseline="wiggle"

plt.legend(loc = 2)
plt.show()
```
