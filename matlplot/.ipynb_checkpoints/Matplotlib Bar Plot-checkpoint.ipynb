```python
x = ["Python", "c", "C+", "Java"]
y = [85, 70, 60, 82]
plt.xlabel("Languages", fontsize = 15)
plt.ylabel("NUmbers", fontsize = 15)
plt.title("WS Cube", fontsize = 15)
c =["y", "b", "m", "g"]
#plt.bar(x,y, width = 0.4, color="yellow")
#plt.bar(x,y, width = 0.4, color=c, align="edge")
plt.bar(x,y, width = 0.4, color=c, align="center")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python", "c", "C+", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages", fontsize = 15)
plt.ylabel("NUmbers", fontsize = 15)
plt.title("WS Cube", fontsize = 15)

plt.bar(p,y, width, color="y", edgecolor = "r", label="popularity")
plt.bar(p1,z, width, color="r", edgecolor = "r", label="popularity1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python", "c", "C+", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages", fontsize = 15)
plt.ylabel("NUmbers", fontsize = 15)
plt.title("WS Cube", fontsize = 15)

plt.bar(p,y, width, color="y", edgecolor = "r", label="popularity")
plt.bar(p1,z, width, color="r", edgecolor = "r", label="popularity1")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["Python", "c", "C+", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages", fontsize = 15)
plt.ylabel("NUmbers", fontsize = 15)
plt.title("WS Cube", fontsize = 15)

plt.barh(p,y, width, color="y", edgecolor = "r", label="popularity")
plt.barh(p1,z, width, color="r", edgecolor = "r", label="popularity1")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Python", "c", "C+", "Java"]
y = [85, 70, 60, 82]

z = [20, 30, 40, 50]
plt.xlabel("Languages", fontsize = 15)
plt.ylabel("NUmbers", fontsize = 15)
plt.title("WS Cube", fontsize = 15)

plt.barh(x,y,  color="y", edgecolor = "r", label="popularity")
plt.barh(x,z,  color="r", edgecolor = "r", label="popularity1")
plt.legend()
plt.show()
```
