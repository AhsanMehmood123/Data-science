# Pie Chart

```python
import matplotlib.pyplot as plt
y = ["C", "C++", "java", "Python"]
x = [10,20,30,40]

plt.pie(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
y = ["C", "C++", "java", "Python"]
x = [10,20,30,40]
ex = [0.2,0.0,0.0,0.0]

plt.pie(x,labels=y, explode = ex) # explode

plt.show()
```
