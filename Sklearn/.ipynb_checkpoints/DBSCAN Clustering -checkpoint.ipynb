```python
# Problem statment
'''Imagine you are working for an e-commerce company and the marketing team wants to target customer more 
effectively by creating segments based on their purchasing behvior. However, traditional segmentation methods are 
not providing satisfactory result....'''
#solution:(Density based spatial clustering of application with noise)
# use DBSCAN to identify clusters of customer with similar purchasing patterns.
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
data = np.random.rand(100,2)
```

```python
scalar = StandardScaler()
data_scaled = scalar.fit_transform(data)
```

```python
# applied dbscan

dbscan = DBSCAN(eps=0.3, min_samples=5)

cluster =  dbscan.fit_predict(data_scaled)
```

```python
#visualize
plt.scatter(data[:,0],data[:,1],c=cluster,cmap='viridis', marker='o',s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Freature 1')
plt.ylabel('Freature 2')
plt.show()
```

```python
result_df = pd.DataFrame({'Feature_1':data[:,0], 
                         'Feature_2': data[:,1],
                         'Cluster': cluster
                         })
```

```python
print(f"Numer of clusters: ",len(np.unique(cluster)))

print("Size of each cluster")

print(result_df['Cluster'].value_counts())
```

```python
#model testing user input
user_input = np.array([[0.6,0.7]])
user_input_Scaled = scalar.transform(user_input)
user_cluster =  dbscan.fit_predict(user_input_Scaled)
print(f"user input belong to cluster: {user_cluster}")
```
