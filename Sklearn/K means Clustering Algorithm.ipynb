```python
#problem statment
'''A Mall wants to understand its customer base better to improve its market strategy. 
The goal is cluster customers into group based on their annual income and spending score, 
so the marketing team can tailor their campaigns to each cluster's preferences and behavior'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# generate data set
np.random.seed(42)
data = {'Annual_income': np.random.randint(30000,100000,100),
       'Spending_score': np.random.randint(1,100,100)}
df = pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df['Annual_income'],df['Spending_score'])
plt.title('Customer Data - Annual income vs Spending score')
plt.xlabel('Annual income')
plt.ylabel('Spending score')
plt.show()
```

```python
X = df.values
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
print(df)
```

```python
plt.scatter(df['Annual_income'],df['Spending_score'],c=df['Cluster'],cmap="rainbow")
plt.title('Kmean Clustering - Annual income and Spending score')
plt.xlabel('Annual income')
plt.ylabel('Spending score')
plt.show()
```

```python
#User input
user_input = {'Annual_income': [75000], 'Spending_score': [70]}
user_df = pd.DataFrame(user_input)
user_cluster = kmeans.predict(user_df)
print(f"the user belong two cluster {user_cluster[0]}")
```
